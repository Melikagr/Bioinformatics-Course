# The First Wave - The Protein Sequence and Structure Wave


## ğŸ§± 1. Protein Sequencing & Alignment (1950sâ€“1970s)

In **1955**, *Frederick Sanger* sequenced bovine insulin â€” the first protein ever decoded.  
This moment marked the birth of *biological sequence data*.

As more proteins were sequenced, researchers needed to compare new sequences with known ones:

- Are these proteins similar?
- Do they share structure or function?

This led to foundational algorithms such as:

### ğŸ”¹ **Needlemanâ€“Wunsch Algorithm (1970)**
A dynamic programming method for pairwise sequence alignment:

- rewards matches  
- penalizes mismatches & gaps  
- finds the optimal global alignment  

**Why alignment mattered:**  
Proteins with similar sequences usually share:

- similar **3D structure**  
- similar **biological function**

Sequence comparison became the first computational tool of molecular biology.

---

## ğŸ§¬ 2. Structural Databases (1970sâ€“1990s)

As more structures were solved experimentally, a centralized resource was needed.

### ğŸ”¹ **Protein Data Bank (PDB, 1973)**  
<img src="https://cdn.rcsb.org/rcsb-pdb/v2/about-us/Logo/rcsb-pdb-logo.png" width="750">

Created to store experimentally determined structures.

Today, PDB contains **170,000+** macromolecular structures.

This enabled:

- global access to protein structures  
- large-scale comparison of sequence/structure relationships  
- training data for future structure prediction algorithms  

---

## ğŸ” 3. BLAST & Fast Sequence Search (1990s)

As sequencing exploded, pairwise comparison became too slow.  
What biology needed was a **search engine**.

### ğŸ”¹ **BLAST (Basic Local Alignment Search Tool)**  
<img src="https://microbenotes.com/wp-content/uploads/2022/07/Basic-Local-Alignment-Search-Tool-BLAST.jpg" width="750">

A revolution in biological search.

Instead of aligning a query with every known sequence, BLAST:

- finds promising local matches first  
- refines only the most likely alignments  

This enabled:

- rapid homology detection  
- gene annotation  
- inference of structure/function from similar sequences

---

## ğŸ§  4. Structure Prediction & CASP (1990sâ€“2010s)

By the early 1990s, scientists asked:

> â€œCan we predict the 3D structure of a protein from its sequence alone?â€

Experimental methods like crystallography were slow.  
Many computational methods were proposed â€” some good, many unreliable.

To objectively evaluate progress, an international blind competition was created.

### ğŸ”¹ **CASP (Critical Assessment of Structure Prediction), 1994**

A global challenge held every 2 years:

1. Participants predict structures for proteins whose experimental structure is not yet public.  
2. Predictions are compared to the true structures using metrics like **GDT (Global Distance Test)**.

CASP revealed:

- the true difficulty of structure prediction  
- a few consistently strong tools (e.g., Rosetta, HHpred)  
- slow but steady progress over two decades

# ğŸŒ¿ The Second Wave of Bioinformatics â€” The Gene Expression Era

The second major evolution in bioinformatics came from a shift in focus:  
**from sequences to expression.**  
Instead of asking *â€œWhat genes exist?â€*, scientists began asking:

> *â€œWhich genes are active, and under what conditions?â€*

This wave was powered by technologies that made it possible to measure gene expression at increasingly massive scales.

---

## ğŸ§ª 1. Early Beginnings â€” Northern Blot (1977)

The story begins in **1977** with the invention of the **Northern blot**,  
a technique that allowed scientists to measure the expression level of **one gene at a time**.

Characteristics of Northern blot:

- labor-intensive  
- low throughput  
- requires substantial experimental skill  
- measures only a single transcript per experiment  

Yet it introduced a revolutionary idea:

> Gene expression can be quantified.

This idea paved the way for much larger and more ambitious technologies.

---

## ğŸ§¬ 2. The Rise of Microarrays (1995â€“2000)

<img src="https://assets.technologynetworks.com/production/dynamic/images/content/368539/innovations-and-versatility-in-modern-microarray-technology-368539-960x540.jpg?cb=12075259" width="500">

A massive leap occurred in **1995** at Stanford University with the development of **DNA microarrays**.

For the first time, researchers could measure **hundreds or thousands of genes simultaneously** using:

- a membrane or glass slide  
- thousands of immobilized DNA probes  
- fluorescently labeled RNA or cDNA from samples  

### ğŸ”¹ How microarrays worked

1. Collect two samples (e.g., **normal** vs. **cancer** cells).  
2. Extract RNA from each.  
3. Convert RNA to cDNA and label each sample with a different fluorescent dye.  
4. Hybridize both samples onto the same array.  
5. Read the signal:

   - **Dark spots** â†’ low expression in both  
   - **Yellow** â†’ high expression in both  
   - **Green** â†’ higher in normal cells  
   - **Red/Blue** â†’ higher in cancer cells  

This technology allowed researchers to ask questions that were impossible before:

- Which genes change in disease?
- Which pathways are activated?
- How do tissues differ at the molecular level?

---

## ğŸ§« 3. Commercial High-Density Arrays

Initially, microarrays were â€œhome-madeâ€:  
labs printed their own cDNA probes onto membranes.

Problems:

- batch effects  
- artifacts  
- inconsistent manufacturing  

This led to industrial solutions such as **Affymetrix GeneChips**, containing:

- up to **six million probes**  
- covering **~20,000 genes**  
- highly reproducible  
- standardized across labs  

These commercial arrays accelerated the field and enabled large, multicenter studies.

### ğŸ”¹ Problem
Two leukemias â€” **ALL (Acute Lymphoblastic Leukemia)** and **AML (Acute Myeloid Leukemia)** â€” look very similar under the microscope.

### ğŸ”¹ Approach

2. Used microarrays to measure thousands of genes per patient.  
3. Identified **marker genes** distinguishing the two diseases.  

### ğŸ”¹ Impact
For the first time, gene expression could be used for:

- accurate molecular diagnosis  
- guiding treatment  
- enabling precision medicine

# ğŸ§¬ The Third Wave of Bioinformatics â€” The DNA Sequencing Era

The next major revolution in bioinformatics came from the rise of **DNA sequencing**.  
This wave transformed molecular biology from a gene-by-gene discipline into a **genome-scale science**.

It enabled everything from the Human Genome Project to modern single-cell sequencing.

---

## ğŸ§ª 1. Foundations â€” DNA as the Molecule of Life (1953â€“1970s)

The story begins in **1953**, when Watson and Crick described the **double helix**.  
This discovery set the foundation for understanding heredity as a digital, code-like system.

But scientists still had no practical way to *read* the sequence of DNA.

### ğŸ”¹ Recombinant DNA (1972)

In **1972**, molecular biologists developed **recombinant DNA technology**,  
which allowed researchers to:

- cut DNA  
- splice DNA fragments  
- assemble new constructs  

This enabled functional studies of specific genes but **sequencing was still extremely slow**.

---

## ğŸ§¬ 2. Early Sequencing â€” A Single Gene Takes Five Years (1970s)

In the early 1970s, sequencing even a *tiny* gene was a monumental challenge.

A famous example:

> A 1972 PhD thesis by Jerry Rubin reported the sequence of **yeast 5S rRNA**,  
> and the project took **five years of experimental work**.

Today, sequencing that same fragment takes **less than one second**.

But at the time, sequencing methods were primitive and required massive effort.

---

## ğŸ§ª 3. Sanger Sequencing (1977)

In **1977**, Frederick Sanger introduced a breakthrough method for determining DNA sequence.  
Sanger sequencing enabled accurate reading of DNA, but:

- it required **large amounts of DNA**,  
- it was limited in throughput,  
- and it was too slow for large genomes.

This limitation led to one of the most crucial inventions in molecular biology.

---

## ğŸ”¥ 4. PCR â€” The Amplification Revolution (1985)

The **Polymerase Chain Reaction (PCR)** was introduced in **1985**.

PCR allowed researchers to:

- amplify a specific DNA fragment **millions of times**  
- starting even from **a single copy**  

With PCR, scientists finally had enough DNA to feed into Sanger sequencing.

Result:

> DNA sequencing became accessible, routine, and scalable.

This set the stage for large-scale genome sequencing.

---

## ğŸ§­ 5. NCBI â€” Organizing the DNA World (1988)

As DNA and protein sequences accumulated, the U.S. NIH founded the  
**National Center for Biotechnology Information (NCBI)** (1988).

NCBI became the central hub for:

- protein sequences  
- DNA sequences  
- gene expression data  
- BLAST search  
- structural databases  
- taxonomy  
- biomedical literature (PubMed)  
- genome assemblies  

NCBI allowed researchers worldwide to **store, share, compare, and annotate sequence data**.

## ğŸ§¬ 6. The Human Genome Project (1990â€“2003)

In **1990**, NIH launched the ambitious **Human Genome Project (HGP)**.  
It was motivated by rapid improvements in:

- PCR  
- Sanger sequencing  
- automated DNA sequencers  

The international public effort involved the U.S. plus six other countries.

### ğŸ”¹ Strategy: *Clone-by-clone sequencing*
The human genome was likened to a giant encyclopedia:

- Chromosome = volume  
- Regions = chapters  
- Subregions = pages  

Steps:

1. Partition the genome into large fragments  
2. Map each fragment  
3. Subdivide into smaller fragments  
4. Sequence the pieces repeatedly  
5. Align overlapping sequences  
6. Assemble into **contigs**  
7. Join contigs into chromosomal sequences  

This workflow was systematic but **painfully slow**.

---

## âš¡ 7. The Private Challenger â€” Celera Genomics (1997)

In **1997**, Craig Venter announced that his private company, **Celera Genomics**,  
could sequence the human genome **faster and cheaper**.

### ğŸ”¹ Strategy: Whole-genome shotgun sequencing

Instead of:

- dividing the genome  
- mapping  
- sub-mapping  
- sequencing chunk-by-chunk

Celera proposed:

> â€œShred the entire genome into small pieces at once.  
> Sequence everything.  
> Use computational algorithms to assemble the genome from overlapping fragments.â€

This required **massive computational power** and advanced assembly algorithms.  
At first, the public consortium was skeptical â€” until Celera began showing rapid progress.

The competition triggered:

- faster automation  
- improved sequencing machines  
- aggressive scaling  
- better assembly pipelines  

---

## ğŸ 8. The Genome Race â€” Collaboration and Completion

By **2000**, both the public and private groups were racing at full speed.  
The White House eventually facilitated a coordinated joint announcement:

> The first working **draft of the human genome** was completed in **Spring 2000**.

The final refined genome was published in **2003**.

### ğŸ”¹ Fun fact
The reference genome consisted of DNA from **nine individuals**, representing the â€œcommon consensusâ€ sequence.

## ğŸ§¬ 9. After the Genome: Variants, Traits, and Personal Genomics

Completing the human reference genome was not the end â€” it was the beginning.

Researchers now asked:

- How do genomes differ between individuals?  
- How do variants influence traits (eye color, disease risk, drug response)?  
- How do we identify genes and annotate their functions?  

This launched entire fields such as:

- **comparative genomics**  
- **GWAS (Genome-Wide Association Studies)**  
- **functional genomics**  
- **personal genomics**
